## compare all data from two folders

```python
for idx,data_dic in enumerate(data_raw_list_1):
    fig,ax = plt.subplots()
    plt.title("plot latency" + data_dic['name'])

    print(labels[0],data_raw_list_1[idx]['name'])
    if len(data_dic['time']) == 1:
        print("Size = 1 Skip")
    else:
        (latency_avg,latency_max,latency_min,latency_std,data_sub_1) = Latecy_analysis(data_raw_list_1[idx]['time'])
        print("latency_avg: ", latency_avg,"latency_min: ", latency_min,"latency_max: ", latency_max,"latency_std: ",latency_std)

    print(labels[1],data_raw_list_2[idx]['name'])
    if len(data_dic['time']) == 1:
        print("Size = 1 Skip")
    else:    
        (latency_avg,latency_max,latency_min,latency_std,data_sub_2) = Latecy_analysis(data_raw_list_2[idx]['time'])
        print("latency_avg: ", latency_avg,"latency_min: ", latency_min,"latency_max: ", latency_max,"latency_std: ",latency_std)

        ax.plot(data_sub_1, label=labels[0])
        ax.plot(data_sub_2, label=labels[1])

        ax.legend(loc='upper right')
        plt.show()

```

### Jitter Hist

```python
for idx,data_dic in enumerate(data_raw_list_1):
    fig,ax = plt.subplots()
    plt.title("Histogram jitter" + data_raw_list_1[idx]['name'])
    
    print(labels[0],data_raw_list_1[idx]['name'])
    
    if len(data_dic['time']) == 1:
        print("Size = 1 Skip")
    else:
        start = int(len(data_raw_list_1[idx]['time'])*SKIP_PERC)
        end   = len(data_raw_list_1[idx]['time']) - int(len(data_raw_list_1[idx]['time'])*SKIP_PERC)
        (jitter_avg,jitter_max,jitter_min,jitter_std,jitter_sub_1) = Jitter_analysis(data_raw_list_1[idx]['time'][start:end])
        print("len",len(data_dic['time']),"jitter_avg: ", jitter_avg,"jitter_min: ", jitter_min,"jitter_max: ", jitter_max,"jitter_std: ",jitter_std)
    
    print(labels[1],data_raw_list_2[idx]['name'])
    if len(data_dic['time']) == 1:
        print("Size = 1 Skip")
    else:
        start = int(len(data_raw_list_2[idx]['time'])*SKIP_PERC)
        end   = len(data_raw_list_2[idx]['time']) - int(len(data_raw_list_2[idx]['time'])*SKIP_PERC)
        (jitter_avg,jitter_max,jitter_min,jitter_std,jitter_sub_2) = Jitter_analysis(data_raw_list_2[idx]['time'][start:end])
        print("len",len(data_dic['time']),"jitter_avg: ", jitter_avg,"jitter_min: ", jitter_min,"jitter_max: ", jitter_max,"jitter_std: ",jitter_std)

        ax.hist(jitter_sub_1, "auto" , None, fc='none', lw=1.5, histtype='step', label=labels[0])
        ax.hist(jitter_sub_2, "auto" , None, fc='none', lw=1.5, histtype='step', label=labels[1])

        ax.legend(loc='upper right')
        plt.show()
```

### Jitter plot

```python
for idx,data_dic in enumerate(data_raw_list_1):
    fig,ax = plt.subplots()
    plt.title("plot Jitter" + data_dic['name'])

    print(labels[0],data_raw_list_1[idx]['name'])
    start = int(len(data_raw_list_1[idx]['time'])*SKIP_PERC)
    end   = len(data_raw_list_1[idx]['time']) - int(len(data_raw_list_1[idx]['time'])*SKIP_PERC)
    
    print(labels[0],data_raw_list_1[idx]['name'])
    if len(data_dic['time']) == 1:
        print("Size = 1 Skip")
    else:
        (latency_avg,latency_max,latency_min,latency_std,data_sub_1) = Jitter_analysis(data_raw_list_1[idx]['time'][start:end])
        print("latency_avg: ", latency_avg,"latency_min: ", latency_min,"latency_max: ", latency_max,"latency_std: ",latency_std)
    
    print(labels[1],data_raw_list_2[idx]['name'])
    if len(data_dic['time']) == 1:
        print("Size = 1 Skip")
    else:
        (latency_avg,latency_max,latency_min,latency_std,data_sub_2) = Jitter_analysis(data_raw_list_2[idx]['time'][start:end])
        print("latency_avg: ", latency_avg,"latency_min: ", latency_min,"latency_max: ", latency_max,"latency_std: ",latency_std)

        ax.plot(data_sub_1, label=labels[0])
        ax.plot(data_sub_2, label=labels[1])

        ax.legend(loc='upper right')
        plt.show()
```
