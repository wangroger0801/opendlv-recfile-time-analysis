### latency analysis
latency = received time - send time

```python
for data_dic in (data_raw_list):
    print(data_dic['name'])
    if len(data_dic['time']) == 0:
        print("Size = 0 Skip")
    else:
        (latency_avg,latency_max,latency_min,latency_std,data_sub) = Latecy_analysis(data_dic['time'])
        print("latency_avg: ", latency_avg,"latency_min: ", latency_min,"latency_max: ", latency_max)
        print("latency_std: ", latency_std)
        plt.plot(data_sub)
        plt.title("plot latency " + data_dic['name'])
        plt.show()
```

### Jitter
Jitter  = timestamp[t+1] - timestamp[t] 

```python
for data_dic in (data_raw_list):
    print("------------------------")

    print(data_dic['name'])
    if len(data_dic['time']) == 1:
        print("Size = 1 Skip")
    else:
        start = int(len(data_dic['time'])*SKIP_PERC)
        end   = len(data_dic['time']) - int(len(data_dic['time'])*SKIP_PERC)
        (jitter_avg,jitter_max,jitter_min,jitter_std,jitter_sub) = Jitter_analysis(data_dic['time'][start:end])
        print("len",len(data_dic['time']),"jitter_avg: ", jitter_avg,"jitter_min: ", jitter_min,"jitter_max: ", jitter_max)
        print("jitter_std: ",jitter_std)
        plt.hist(jitter_sub, bins='auto') 
        plt.title("Histogram jitter " + data_dic['name'])
        plt.show()
        plt.plot(jitter_sub)
        plt.title("plot jitter " + data_dic['name'])
        plt.show()
        #print(np.argmax(jitter_sub))
        #print(data_dic['time'][np.argmax(jitter_sub)])
        #print(data_dic['time'][np.argmax(jitter_sub)])
        #print(np.argmax(jitter_sub)+1)
        #print(data_dic['time'][np.argmax(jitter_sub)+1])
```
